```python
copy['date_added'] = pd.to_datetime(copy['date_added'])
copy['day_added'] = copy['date_added'].dt.day
copy['month_added'] = copy['date_added'].dt.month
copy['year_added'] = copy['date_added'].dt.year
copy['year_added'].astype(int)
copy['day_added'].astype(int)
```

```python
sns.countplot(data=netflix_data, x='type', palette='colorblind')
plt.title('Type')
plt.show()
```

```python
sns.countplot(data=netflix_data, x='rating', order=netflix_data['rating'].value_counts().index, palette='deep')
plt.title('Rating')
plt.xticks(rotation=90, ha= 'right')
plt.show()
```

```python
sns.countplot(data=netflix_data, x='rating', hue='type')
plt.xticks(rotation=90)
plt.title('Relationship Between Type and Rating')
plt.legend()
plt.show()
```

```python
labels = ['Movie', 'Tv Show']
size = netflix_data['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]

plt.figure(figsize=(9, 9))
plt.pie(size, labels=labels, colors=colors, explode=explode, shadow = True, startangle=90, autopct='1%.1f%%')
plt.title('Distribution of Type', fontsize=20)
plt.legend()
plt.show()
```

```python
netflix_data['rating'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(10,8))
plt.show()
```
